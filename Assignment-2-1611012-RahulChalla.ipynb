```python
from nltk.tokenize import sent_tokenize,word_tokenize
import math
import nltk
```

```python
file = open("speeches.txt",encoding="utf-8")
text = file.read()
text = text[text.index('T'):]
sentences = sent_tokenize(text)
trainsents = []
testsents = []
for i in sentences[:math.floor(0.8*len(sentences))]:
    trainsents.append("<s> "+i+" </s>")
for i in sentences[math.floor(0.8*len(sentences)):]:
    testsents.append("<s> "+i+" </s>")
```

```python
bid = {}
trid = {}
quad = {}
for i in trainsents:
    bigs = nltk.bigrams(word_tokenize(i))
    for big in bigs:
        keyy = "|".join(list(big[:-1]))
        vall = big[-1]
        if(keyy not in bid.keys()):
            bid[keyy] = [{vall:2},2]
        else:
            if(vall in bid[keyy][0]):
                bid[keyy][0][vall] = bid[keyy][0][vall]+1
                bid[keyy][1] = bid[keyy][1] + 1
            else:
                bid[keyy][0][vall] = 2
                bid[keyy][1] = bid[keyy][1] + 1
```

```python
for i in trainsents:
    bigs = nltk.trigrams(word_tokenize(i))
    for big in bigs:
        keyy = "|".join(list(big[:-1]))
        vall = big[-1]
        if(keyy not in bid.keys()):
            trid[keyy] = [{vall:2},2]
        else:
            if(vall in trid[keyy][0]):
                trid[keyy][0][vall] = trid[keyy][0][vall]+1
                trid[keyy][1] = trid[keyy][1] + 1
            else:
                trid[keyy][0][vall] = 2
                trid[keyy][1] = 2
```

```python
for i in trainsents:
    bigs = nltk.ngrams(word_tokenize(i),4)
    for big in bigs:
        keyy = "|".join(list(big[:-1]))
        vall = big[-1]
        if(keyy not in bid.keys()):
            quad[keyy] = [{vall:2},2]
        else:
            if(vall in quad[keyy][0]):
                quad[keyy][0][vall] = quad[keyy][0][vall]+1
                quad[keyy][1] = quad[keyy][1] + 1
            else:
                quad[keyy][0][vall] = 2
                quad[keyy][1] = 2
```

```python
resmle = []
for i in testsents:
    print(i)
    testwords = word_tokenize(i)
    mle = 1
    for j in range(3,len(testwords)):
        try:
            mle = mle * quad['|'.join(testwords[j-3:j])][0][testwords[j]]*1.0/quad['|'.join(testwords[j-3:j])][1]
        except:
            mle = mle * 1.0/quad['|'.join(testwords[j-3:j])][1]
        finally:
            mle = mle * 1.0
    resmle.append(mle)
```
